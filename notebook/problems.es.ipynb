## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import random

e=0
for i in range(1000):
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)
    s = d1+d2

    if s > 7:
        e=e+1
p = (e/1000)*100
print('La probabilidad experimental de que tirando dos dados la suma resultante sea mayor que 7 es del {:.2f}%'.format(p))
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
# by experimental probability
win=0
win_white=0
win_red=0
win_green=0
for i in range(1000):
    sel = random.choices(ball_box, k=5)
    white = sel.count('White')
    red = sel.count('Red')
    green = sel.count('Green')

    if white == 3 and red == 2:
        win = win+1
    if white == 5:
        win_white = win_white+1
    if red == 5:
        win_red = win_red+1
    if green == 5:
        win_green = win_green+1
p2 = (win/1000)*100
pw = (win_white/1000)*100
pr = (win_red/1000)*100
pg = (win_green/1000)*100
pall = pw+pr+pg

print('Empirical probability of take 3 white balls and 2 red balls: {:.2f}%'.format(p2))
print('Empirical probability of take 5 white balls: {:.2f}%'.format(pw))
print('Empirical probability of take 5 red balls: {:.2f}%'.format(pr))
print('Empirical probability of take 5 green balls: {:.2f}%'.format(pg))
print('Empirical probability of take 5 balls of the same color: {:.2f}%'.format(pall))
```

```python
# by joint probability
import numpy as np
import math as m
# The way the ball is returned after noting its color, is not conditional probability
# When returning the ball, subsequently drawing 5 balls is equivalent to having 5 identical 
# boxes and take one ball from each simultaneously. 
# Therefore, the probability to be calculated is the joint probability

# probability of get one ball of each color
prob_white=10/60
prob_red=20/60
prob_green=30/60

# Like the white balls are indestingible and red balls too, the number of combinations of 
# 3 white balls and 2 red balls are given by:

good_comb=m.factorial(5)/(m.factorial(3)*m.factorial(2))
# there is only one good combination forget all balls of the same color:
good_comb_samecolor = 1

# the probability of get one valid subset form the box:
prob_subset = (prob_white**3)*(prob_red**2)
prob_subset_w = (prob_white**5)
prob_subset_r = (prob_red**5)
prob_subset_g = (prob_green**5)

# probability of get any of good combination from the box:
p3 = (prob_subset*good_comb)*100
p3w = (prob_subset_w*good_comb_samecolor)*100
p3r = (prob_subset_r*good_comb_samecolor)*100
p3g = (prob_subset_g*good_comb_samecolor)*100
p4 = p3w+p3g+p3g

print('Theoretical probability of take 3 white balls and 2 red balls: {:.2f}%'.format(p3))
print('Theoretical probability of take 5 white balls: {:.2f}%'.format(p3w))
print('Theoretical probability of take 5 red balls: {:.2f}%'.format(p3r))
print('Theoretical probability of take 5 green balls: {:.2f}%'.format(p3g))
print('Theoretical probability of take 5 balls of the same color: {:.2f}%'.format(p4))
```
